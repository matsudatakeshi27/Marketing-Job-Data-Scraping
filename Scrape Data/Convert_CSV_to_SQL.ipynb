```python
import pandas as pd
import mysql.connector
```

```python
data = pd.read_csv("linkedin-jobs.csv")
data
```

```python
data = data.fillna('')
```

```python
#Connect to SQL Server
mydb = mysql.connector.connect(
    host = "localhost",
    user = "root",
    password = "",
    database = "job_scraping",
    charset='utf8mb4'
)

cursor = mydb.cursor()

# cursor.execute('''
#                 CREATE TABLE Marketing_Job (
#                     Job_title varchar(200),
#                     Company_name varchar(100),
#                     Location varchar(100),
#                     Apply varchar(300),
#                     Job_description text,
#                     Date varchar(50),
#                     Company_link varchar(300)
#                 )
# ''')

query = "INSERT INTO Marketing_Job (Job_title, Company_name, Location, Apply, Job_description, Date, Company_link) VALUES (%s, %s, %s, %s, %s, %s, %s)"

# Iterate over each row in the DataFrame and insert the values into the table
for _, row in data.iterrows():
    values = (
        row['Title'],
        row['Company'],
        row['Location'],
        row['Apply'],
        row['Job Description'],
        row['Date'],
        row['Company Link']
    )
    cursor.execute(query, values)

# Commit the changes to the database
mydb.commit()

# Close the cursor and database connection
cursor.close()
mydb.close()
```
